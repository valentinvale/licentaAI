```python
import cv2
import matplotlib.pyplot as plt

# Assuming 'contours' is a list of contours obtained from image processing
# (e.g., using cv2.findContours)

# Example: Check if contours list is not empty
if contours and len(contours[0]) > 0:
    # Assuming you want to process the first contour in the list
    firstContour = contours[0].astype(int)

    # Check if the firstContour array is not empty
    if len(firstContour) > 0:
        startRow = min(firstContour[:, 0])
        endRow = max(firstContour[:, 0])
        startCol = min(firstContour[:, 1])
        endCol = max(firstContour[:, 1])
    else:
        startRow, endRow, startCol, endCol = 0, 0, 0, 0

    # Display the region of interest
    fig, ax = plt.subplots()
    ax.imshow(testImage[startRow:endRow, startCol:endCol], cmap=plt.cm.gray)
    ax.axis('image')
    ax.set_xticks([])
    ax.set_yticks([])

    plt.show()
else:
    print("No contours found.")
```
